## basic Imports functions

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
heart_dis_df = pd.read_csv("C:/Users/Safeer/Documents/MachineLearning/heart-disease.csv")
```

```python
heart_dis_df
```

```python
#create X ( Feature Matrix)
X_data = heart_dis_df.drop('target', axis=1)

#creating Y ( Predection Matrix)
y = heart_dis_df['target']
```

```python
X_data , y
```

### chosing the models to classifey

```python
clf = RandomForestClassifier ()

clf.get_params()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_data,y, test_size=.2)
x_train
```

```python
clf.fit(x_train,y_train)
```

```python
y_label = clf.predict(x_test)
y_label
```

```python
clf.score(x_test, y_test)
```

```python
from  sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_label))
print (confusion_matrix(y_test, y_label))
print (accuracy_score(y_test, y_label))
```

## making my custom sting to intiger convertor in car_sales dataframe

```python
car_sales_make_keys = car_sales.Make.value_counts().keys()
```

```python
cs_make_dic = dict()
for i in range(len(car_sales_make_keys)):
    cs_make_dic [car_sales_make_keys[i]] = i
    
cs_make_dic
```

```python
car_sales['Make'] = car_sales['Make'].replace(cs_make_dic, regex=True)
car_sales
```

```python
car_sales.Colour.unique()
```

```python
cs_colour_dic = dict()
car_sales_colours_keys = car_sales.Colour.unique()
for i in  range(len(car_sales_colours_keys)) :
    cs_colour_dic[car_sales_colours_keys[i]] = i

cs_colour_dic
```

```python
car_sales.Colour = car_sales.Colour.replace(cs_colour_dic, regex=True)
car_sales
```
