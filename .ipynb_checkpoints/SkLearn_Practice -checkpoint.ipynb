Using sklearn model to fill the missing data

```python
car_miss_df = pd.read_csv('C:/Users/Safeer/Documents/MachineLearning/car-sales-extended-missing-data.csv')
car_miss_df.head()
```

```python
car_miss_df.dropna(subset=['Price'], inplace= True)
car_miss_df.isnull().sum()
```

```python
X = car_miss_df.drop(columns='Price')
y = car_miss_df['Price']
```

```python
# import missing values
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#fill categorical valus using imputer
catag_imputer = SimpleImputer(strategy='constant',fill_value='missing')
door_imputer  = SimpleImputer(strategy='constant',fill_value=4)
num_imputer   = SimpleImputer(strategy='mean')

#define columns
cat_features = ['Make','Colour']
door_features= ['Doors']
num_features = ['Odometer (KM)']

#Create an imputer
imputer = ColumnTransformer ([
    ('cat_imputer', catag_imputer, cat_features),
    ('door_imputer',door_imputer,  door_features),
    ('num_imputer', num_imputer,   num_features)
])

filled_X = imputer.fit_transform(X)
filled_X
```

```python
new_X = pd.DataFrame(filled_X, columns=['Make','Colour','Door','Odometer (KM)'])
new_X
```

```python
new_X.isna().sum()
```

```python
new_X.info()
```

```python
new_X['Odometer (KM)'] = new_X['Odometer (KM)'].astype(int)
new_X['Door'] = new_X['Door'].astype(int)
```

```python
new_X.info()
```

```python
new_one_hot_X = pd.get_dummies (new_X)
new_one_hot_X
```

```python
y_miss.info()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# Split data in test and train test
X_train, X_test, y_train, y_test  = train_test_split ( new_one_hot_X, y_miss, test_size= 0.2 )
```

```python
print (X_train.shape)
print (y_train.shape)
```

```python
# Model declearation and model fitting
rig = Ridge()

rig.fit( X_train, pd.DataFrame(y_train))

rig.score(X_test, y_test)
```

```python
from sklearn.svm import SVR
svr = SVR()

svr.fit( X_train, y_train)
svr.score(X_test, y_test)
```

```python
from sklearn.linear_model import SGDRegressor

sdgr1 = SGDRegressor(loss="squared_error")
sdgr2 = SGDRegressor(loss="huber")
sdgr3 = SGDRegressor(loss="epsilon_insensitive")

print ("Ridge ",rig.score(X_test, y_test))

print ("SVR ",svr.score(X_test, y_test))

sdgr1.fit( X_train, y_train)
print ("SGDRegressor with square loss",sdgr1.score(X_test, y_test))

sdgr2.fit( X_train, y_train)
print ("SGDRegressor with huber  loss",sdgr2.score(X_test, y_test))

sdgr3.fit( X_train, y_train)
print ("SGDRegressor with huber  loss",sdgr3.score(X_test, y_test))

```

The best fit is Rodge module in sklearn.linearmodel
