```python
items_dic =    {
    'peanut' : 15,
    'cocunet': 25,
    'almond' : 35
}

#print (items_dic.keys(), items_dic.values())

fig, ax = plt.subplots(figsize = (5,5))

ax.bar (items_dic.keys(), items_dic.values(), width= .15)
ax.set(title="Items lists", ylabel= " Price per Kg in USD")
plt.show()
```
