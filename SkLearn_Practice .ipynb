## making my custom sting to intiger convertor in car_sales dataframe

```python
car_sales_make_keys = car_sales.Make.value_counts().keys()
```

```python
cs_make_dic = dict()
for i in range(len(car_sales_make_keys)):
    cs_make_dic [car_sales_make_keys[i]] = i
    
cs_make_dic
```

```python
car_sales['Make'] = car_sales['Make'].replace(cs_make_dic, regex=True)
car_sales
```

```python
car_sales.Colour.unique()
```

```python
cs_colour_dic = dict()
car_sales_colours_keys = car_sales.Colour.unique()
for i in  range(len(car_sales_colours_keys)) :
    cs_colour_dic[car_sales_colours_keys[i]] = i

cs_colour_dic
```

```python
car_sales.Colour = car_sales.Colour.replace(cs_colour_dic, regex=True)
car_sales
```

```python
car_sales.dtypes
```

```python
# declaring the test and train data from car sale data frama
x = car_sales.drop('Price', axis=1)
y = car_sales.Price
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train
```

```python
y_train
```

```python
clf1 = RandomForestClassifier ()
```

```python
clf1.fit(x_train, y_train)
```

```python
y_predict = clf1.predict(x_test)
y_predict
```

```python
clf1.score(x_test,y_test)
```

```python
confusion_matrix(y_test, y_predict)
```
